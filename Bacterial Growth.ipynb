```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

SMALL_SIZE = 16
MEDIUM_SIZE = 18
BIGGER_SIZE = 20

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
control = np.array([58,46,57])
glucose = np.array([51,62,57])
ketone = np.array([53,62,60])
data = np.array([control,glucose,ketone])
print(data.shape)
mean = np.mean(data,axis=1)
sd = np.std(data,axis=1)
```

```python
plt.figure(figsize=(7,4))
plt.boxplot([control,glucose,ketone],labels=["Control","Glucose","Ketone"])
plt.ylabel("Percent Bacterial Growth")
plt.savefig("bacteria.pdf")
```

```python
fig, ax = plt.subplots(figsize=(7,4))
for datacolumn in data.T:
    line, = ax.plot(["Control","Glucose","Ketone"],datacolumn,"or")
line.set_label("measured")
plt.errorbar(["Control","Glucose","Ketone"],mean,yerr=sd,fmt="s",label="mean+std")
plt.legend(loc=4)
plt.ylabel("Percent Bacterial Growth")
plt.savefig("bacteria2.pdf")
```
